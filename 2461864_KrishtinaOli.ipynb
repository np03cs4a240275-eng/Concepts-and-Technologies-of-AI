```python
#3.1
#6
# Create HDI Category column based on UNDP thresholds
def classify_hdi(value):
    if value < 0.550:
        return "Low"
    elif value < 0.700:
        return "Medium"
    elif value < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df["HDI_Category"] = hdi_2022_df["hdi"].apply(classify_hdi)
# Check category distribution
hdi_2022_df["HDI_Category"].value_counts()
# Display sample rows to confirm new column
hdi_2022_df[["country", "hdi", "HDI_Category"]].head(10)


```

```python
#4
#4.2
#3

# 1. Detect outliers using 1.5 Ã— IQR rule

# HDI IQR
Q1_hdi = HDI_problem1B['hdi'].quantile(0.25)
Q3_hdi = HDI_problem1B['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi

lower_hdi = Q1_hdi - 1.5 * IQR_hdi
upper_hdi = Q3_hdi + 1.5 * IQR_hdi

# GNI per Capita IQR
Q1_gni = HDI_problem1B['gross_inc_percap'].quantile(0.25)
Q3_gni = HDI_problem1B['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni

lower_gni = Q1_gni - 1.5 * IQR_gni
upper_gni = Q3_gni + 1.5 * IQR_gni

# Identify outliers
outliers = HDI_problem1B[
    (HDI_problem1B['hdi'] < lower_hdi) |
    (HDI_problem1B['hdi'] > upper_hdi) |
    (HDI_problem1B['gross_inc_percap'] < lower_gni) |
    (HDI_problem1B['gross_inc_percap'] > upper_gni)
]


# 2. Scatter plot with outliers highlighted


plt.figure(figsize=(8, 6))

# Normal points
plt.scatter(
    HDI_problem1B['gross_inc_percap'],
    HDI_problem1B['hdi'],
    label='Normal',
)

# Outliers
plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    label='Outliers'
)

plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("HDI vs GNI per Capita with Outliers Highlighted")
plt.legend()
plt.show()
```

```python
#5
#5.7
# Function to detect outliers using 1.5 Ã— IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)]

# SOUTH ASIA OUTLIERS
sa_hdi_outliers = detect_outliers(HDI_SouthAsia_2020_2022, 'hdi')
sa_gni_outliers = detect_outliers(HDI_SouthAsia_2020_2022, 'gross_inc_percap')

# Scatter plot: South Asia
plt.figure(figsize=(7,5))
plt.scatter(
    HDI_SouthAsia_2020_2022['gross_inc_percap'],
    HDI_SouthAsia_2020_2022['hdi'],
    label='Normal'
)
plt.scatter(
    sa_gni_outliers['gross_inc_percap'],
    sa_gni_outliers['hdi'],
    label='Outliers'
)
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("South Asia: HDI vs GNI per Capita (Outliers Highlighted)")
plt.legend()
plt.show()

# MIDDLE EAST OUTLIERS
me_hdi_outliers = detect_outliers(HDI_MiddleEast_2020_2022, 'hdi')
me_gni_outliers = detect_outliers(HDI_MiddleEast_2020_2022, 'gross_inc_percap')

# Scatter plot: Middle East
plt.figure(figsize=(7,5))
plt.scatter(
    HDI_MiddleEast_2020_2022['gross_inc_percap'],
    HDI_MiddleEast_2020_2022['hdi'],
    label='Normal'
)
plt.scatter(
    me_gni_outliers['gross_inc_percap'],
    me_gni_outliers['hdi'],
    label='Outliers'
)
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Middle East: HDI vs GNI per Capita (Outliers Highlighted)")
plt.legend()
plt.show()
```
